```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('classic')
```

# State vs. OOP
Using a simple example

```python
lemonade_sales = [78, 90, 83, 82, 60, 95, 82]
temperature = [84, 88, 87, 85, 78, 82, 83]
price = [1.00, 1.00, 1.25, 1.00, 1.50, 1.50, 1.25]
weekend = [False, False, False, False, False, True, True]
lemonade_df = pd.DataFrame(zip(lemonade_sales, temperature, price, weekend))
lemonade_df.columns = ["total_sales", "temp_f", "price", "weekend"]
lemonade_df.price = lemonade_df.price.apply(lambda x: "{:.2f}".format((x))).apply(lambda x:float(x))
lemonade_df
```

```python
lemonade_df.info()
```

```python
# we can use the scatter method directly on the plt module
plt.scatter(lemonade_df["temp_f"], lemonade_df["total_sales"])
```

```python
# or we can create the same effect by first explicitly creating a Figure and Axes
fig, ax = plt.subplots()
ax.scatter(lemonade_df["temp_f"], lemonade_df["total_sales"])
```

```python
# It's very important to label your plots! 
fig, ax = plt.subplots()
ax.scatter(lemonade_df["temp_f"], lemonade_df["total_sales"])
ax.set_title("Lemonade Sales")
ax.set_xlabel("Temperature (Fahrenheit)")
ax.set_ylabel("Sales (Dollars)")
```

```python
# Here is an example of how you can make a plot nicer to look at 
# (this plot could be a lot prettier but it's an improvement over the defaults!)

fig, ax = plt.subplots() # create one subplot
color = ["week"]
ax.scatter(lemonade_df["temp_f"], # independent variable
           lemonade_df["total_sales"], # dependent variable
           # use a list comprehension to highlight weekend sales
           c=["#999999" if not is_weekend else "#FFFF99" for is_weekend in lemonade_df["weekend"]],
           s=lemonade_df["price"]*500 # adjust size according to price           
           )
ax.set_title("Lemonade Sales") # set the title
ax.set_xlabel("Temperature (Fahrenheit)") # label the x-axis
ax.set_ylabel("Sales (Dollars)") # label the y-axis
ax.set_facecolor("#CCEEFF") # change the background color of the Axes object
ax.grid(True) # create a grid on the Axes object
```
